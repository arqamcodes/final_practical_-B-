```python
# Part A

# 1
import pandas as pd
df = pd.read_csv('student_performance.csv')

df.head()


# 2
df.columns = ['Hours_Studied', 'Math_Score', 'Passed', 'Gender', 'Roll_No', 'English_Score']


df = df[['Roll_No', 'Gender', 'Math_Score', 'English_Score', 'Hours_Studied', 'Passed']]
df.head()


```

```python
# Part B
# 1

df['efficiency'] = (df['Math_Score'] + df['English_Score']) / df['Hours_Studied']
df.head()
```

```python
# PartB
# 2
Q1 = df['efficiency'].quantile(0.25)
Q3 = df['efficiency'].quantile(0.75)

IQR = Q3 - Q1


lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR

df_no_outliers = df[(df['efficiency'] >= lower_fence) & (df['efficiency'] <= upper_fence)]

rows_removed = df.shape[0] - df_no_outliers.shape[0]
print("Number of rows removed:", rows_removed)
```
